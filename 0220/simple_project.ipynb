# Project1 : load_digits

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.metrics import roc_curve,roc_auc_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier, LogisticRegression
from xgboost import XGBClassifier
import seaborn as sns
```

```python
digit = load_digits()

digit_data = digit.data #Feature Data 지정
print(digit_data.shape)

digit_label = digit.target # Label Data 지정
print(digit_label.shape)

print(digit.target_names) # target_names
```

```python
print(digit.DESCR)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
for i in range(10):
    plt.subplot(2, 5, i+1) # 10개의 이미지를 2행 5열로 표시
    plt.imshow(digit.data[i].reshape(8, 8), cmap='gray') 
    plt.axis('off') # 축을 보이지 않게 해줍니다. 
plt.show()
```

```python
ax = sns.histplot(digit_label, discrete=True, shrink =0.8)
ax.bar_label(ax.containers[0]);
```

```python
# Data split
X_train, X_test, y_train, y_test = train_test_split(digit_data,
                                                    digit_label,
                                                    test_size=0.2,
                                                    random_state=15)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming y_train and y_test contain the labels of your training and testing data
# Replace y_train and y_test with your actual labels
# Assuming y_train and y_test are lists or NumPy arrays

plt.figure(figsize=(10, 6))

ax = sns.histplot(y_train, discrete=True, shrink=0.8, color='skyblue', label='Train Data')
ax1 = sns.histplot(y_test, discrete=True, shrink=0.8, color='orange', label='Test Data')

# Add bar labels for both histograms
for container in ax.containers:
    ax.bar_label(container)
for container in ax1.containers:
    ax1.bar_label(container)

plt.xlabel('Labels')
plt.ylabel('Frequency')
plt.title('Histogram of Train and Test Labels')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Decision Classifier

```python
decision_tree = DecisionTreeClassifier(random_state=15)
decision_tree.fit(X_train, y_train) # 의사결정나무 모델로 학습
y_pred = decision_tree.predict(X_test) # 테스트 결과 예측

accuracy = accuracy_score(y_test, y_pred)
print(classification_report(y_test,y_pred))
```

## RandomForest

```python
random_forest = RandomForestClassifier(random_state=32) # RandomForest분류기 객체를 생성
random_forest.fit(X_train, y_train) # 훈련
y_pred = random_forest.predict(X_test) # 예측

print(classification_report(y_test, y_pred)) 
```

## Support Vector Machine

```python
svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

## SGD

```python
sgd_model = SGDClassifier()
sgd_model.fit(X_train,y_train)
y_pred = sgd_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

## Logistic Regression

```python
logistic_model = LogisticRegression()
logistic_model.fit(X_train,y_train)
y_pred = logistic_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

# Project3 : load_breast_cancer

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
cancer_data = cancer.data #feature
print(cancer_data.shape)
cancer_label = cancer.target #target
print(cancer_label.shape)
print(cancer.target_names)
```

```python
print(cancer.DESCR)
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(cancer_label, discrete= True, bins=2, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Labels (Malignant / Benign)')
plt.ylabel('Frequency')
plt.title('Histogram of Breast Cancer Labels')
plt.xticks(ticks=[0, 1], labels=cancer.target_names)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer_data,
                                                    cancer_label,
                                                    test_size=0.2,
                                                    random_state=15)
```

```python
Xs_train, Xs_test, ys_train, ys_test = train_test_split(cancer_data,
                                                    cancer_label,
                                                    test_size=0.2,
                                                    random_state=15, stratify=cancer_label)
```

```python
pred_list = []
spred_list = []
```

```python
# Initialize the plot
plt.figure(figsize=(8, 6))
models = ['DT', 'RF', 'SVM', 'SGD', 'LR']
# Loop through each set of predictions and compute ROC curve and AUC score
for i, pred in enumerate(pred_list):
    fpr, tpr, thresholds = roc_curve(y_test, pred)
    auc_score = roc_auc_score(y_test, pred)
    
    # Plot ROC curve
    plt.plot(fpr, tpr, label=f'{models[i]} (AUC = {auc_score:.2f})')

# Plot the random guess line
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random Guess')

# Set labels and title
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')

# Add legend
plt.legend()
plt.grid(True)
plt.show()
```

wine 데이터 또한 multi-label classification문제로 랜덤포레스트와 Logistic Regression이 가장 좋은 성능을 보였는데, 100퍼센트인 점이 조금 의심되었지만, 랜덤시드를 바꿨음에도 100퍼를 보여주었다. 또한, 모델들의 평가는 f1-score 외에 confusion matrix를 통해 시각화를 진행해보았다. 그 외로, Stratified한 train-test split을 적용해 보았는데, SVM의 경우, 성능 향상이 있었지만, 나머지는 오히려 성능이 떨어지는 경향을 보였다.

breast cancer 데이터의 경우, DT와 랜덤포레스트 모델이 강세를 보였으며 이를 f1-score와 auroc 커브를 통해 확인할 수 있었다. stratified하게 샘플링 할 경우, ROC-AUC 커브를 통해 성능 향상을 볼 수 있었다.
